```python
import numpy as np
import matplotlib.pyplot as plt
def mandelbrot(h, w, maxit=20, r=2):
    x = np.linspace(-2.5, 1.5, 4*h+1)
    y = np.linspace(-1.5, 1.5, 3*w+1)
    A, B = np.meshgrid(x, y)
    C = A + B*1j
    z = np.zeros_like(C)
    divtime = maxit + np.zeros(z.shape, dtype=int)
    
    for i in range(maxit):
        z = z**2 + C
        diverge = abs(z) > r 
        div_now = diverge & (divtime == maxit)
        divtime[div_now] = i
        z[diverge] = r
    
    return divtime
print(plt.clf())
print(plt.imshow(mandelbrot(400, 400)))
```

```python
a = np.array([2, 3, 4, 5])
b = np.array([8, 5, 4])
c = np.array([5, 4, 6, 8, 3])
ax, bx, cx = np.ix_(a, b, c)
print(ax)
print(bx)
print(cx)
print(ax.shape, bx.shape, cx.shape)
result = ax + bx + cx
print(result)
print(result[3, 2, 4])
print(a[3] + b[2] + c[4])
```

```python
def ufunc_reduce(ufct, *vectors):
    vs = np.ix_(*vectors)
    r = ufct.identity
    for v in vs:
        r = ufct(r, v)
    return r
```

```python
print(ufunc_reduce(np.add, a, b, c))
```
