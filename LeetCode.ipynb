```python
nums = [4, 1, -1, 2, -1, 2, 3]
k = 2
# def topKFrequent(self, nums: List[int], k: int) -> List[int]
object = {}
for i in nums:
    if i in object:
        object[i] += 1
    else:
        object[i] = 1
for key in object:
    top_k = sorted(object, key=object.get, reverse=True)[:k]
print(object)
print(top_k)

```
